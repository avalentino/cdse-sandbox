### Search DGED-90 DEM tiles intersecting the specified polygon

```python
polygon_wkt = pathlib.Path("polygon.wkt").read_text()
polygon_wkt
```

**Note**: the `items_per_page` option has been set to `50` to have all the items
in the `search_results`.
By default only `20` items are returned.

```python
search_results = dag.search(
    productType="COP_DEM_GLO90_DGED",
    geom=polygon_wkt,
    count=True,
    items_per_page=50,
)
```

```python
search_results
```

### (Optionally) Save search results to file

```python
geojson = search_results.as_geojson_object()
pathlib.Path("search_results_file.geojson").write_text(json.dumps(geojson))
```

### (Optionally) Load search results from file

```python
search_resulst_from_file = dag.deserialize_and_register(
    "search_results_file.geojson"
)
```

### Download the DEM tiles into teh `dem` directory

**Note**: the download requires access to CREODIAS services and hence CREODIAS
authentication credentials.
Credentials can be saved into the EODAG configuration file
(default location`~/.config/eodag/eodag.yml`) or via environment variables as
in teh example below.

**CRITICAL**: in addition to `username` and `password`, CREODIAS also requires
a "TOTP" (one time password) that changes every few seconds.
This makes not handy to use EODAG for DEM downloading.

```python
import os

os.environ["EODAG__CREODIAS__AUTH__CREDENTIALS__USERNAME"] = "xxxxx"
os.environ["EODAG__CREODIAS__AUTH__CREDENTIALS__PASSWORD"] = "xxxxx"
os.environ["EODAG__CREODIAS__AUTH__CREDENTIALS__TOTP"] = "xxxxxx"
```

**Note**: credentials must be already set when the `search` is performed.

```python
dag = eodag.EODataAccessGateway()
search_results = dag.search(
    productType="COP_DEM_GLO90_DGED",
    geom=polygon_wkt,
    count=True,
    items_per_page=50,
)
dag.download_all(search_results, output_dir="dem", extract=False)
```
