```python
from cdseutils.utils import BBox
```

```python
endpoint_url = "https://stac.dataspace.copernicus.eu/v1/"
```

```python
collection = "cop-dem-glo-90-dged-cog"
```

```python
bbox = BBox(
    left=16.50,
    south=40.53,
    right=16.85,
    north=40.72,
)
bbox
```

```python
client = pystac_client.Client.open(endpoint_url)
```

```python
list(client.get_all_collections())
```

```python
search = client.search(
    collections=[collection],
    bbox=bbox,
    # max_items=10,
)
```

```python
list(search.items())
```

```python
dged_search = client.search(
    collections=[collection],
    bbox=bbox,
    # max_items=10,
)
```

```python
items = list(dged_search.items())
items
```
